# STEP 2: Import libraries



```python
#Import libraries
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
```

# STEP 3: Load model

```python
#Load model
import warnings
warnings.filterwarnings("ignore")
tokenizer = AutoTokenizer.from_pretrained("microsoft/DialoGPT-medium")
model = AutoModelForCausalLM.from_pretrained("microsoft/DialoGPT-medium")
```

# STEP 4: Chatbot code

```python
# Chat history
chat_history_ids = None

while True:
    user_input = input("You: ")

    if user_input.lower() in ["exit", "quit"]:
        print("Chatbot: Goodbye!")
        break

    # Encode input
    new_input_ids = tokenizer.encode(user_input + tokenizer.eos_token, return_tensors='pt')

    # Attention mask (FIX)
    attention_mask = torch.ones(new_input_ids.shape, dtype=torch.long)

    # Maintain chat history
    if chat_history_ids is not None:
        input_ids = torch.cat([chat_history_ids, new_input_ids], dim=-1)
        attention_mask = torch.ones(input_ids.shape, dtype=torch.long)
    else:
        input_ids = new_input_ids

    # Generate response
    chat_history_ids = model.generate(
        input_ids,
        attention_mask=attention_mask,
        max_length=1000,
        pad_token_id=tokenizer.eos_token_id,
        do_sample=True,                 # improves response quality
        top_k=50,
        top_p=0.95,
        temperature=0.7
    )

    # Decode response
    bot_response = tokenizer.decode(
        chat_history_ids[:, input_ids.shape[-1]:][0],
        skip_special_tokens=True
    )

    print("Chatbot:", bot_response)
```

# Conclusion

Successfully built a chatbot using Hugging Face Transformers and the DialoGPT model. The chatbot can interact with users and generate dynamic responses based on input.

Since DialoGPT is a generative language model trained on conversational data, its responses may sometimes be unpredictable or less accurate. However, it effectively demonstrates how transformer-based models can simulate human-like conversations.
